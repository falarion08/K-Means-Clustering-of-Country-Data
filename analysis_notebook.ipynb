###Examine null values for each columns

Preprocess null values

```python
# Count null values
data_df.isna().sum()
```

It can be observed that that there are no default values in the dataset, and it is left as blank

```python
# Get index number for each abbreviation that is null
indexList = data_df.loc[data_df['Abbreviation'].isna()].index.tolist()

# For each index update the abbreviation feature with their corresponding alpha 2 code
for index in indexList:
  country_name = data_df.iloc[index]['Country']

  words = country_name.split(' ')
  # Examine each word in the country name if it is a valid country
  for word in words:
    try:
      query = pyc.countries.lookup(word) # Throws an error if the word is not a country name

      # Update the missing abbreviation of country and update the country name with the corresponding name in the pycountries library
      data_df.loc[index,'Country'] = query.name
      data_df.loc[index,'Abbreviation'] = query.alpha_2
      break # End the loop when the valid name is found

    except:
      continue # Continue with the loop if the lookup is an error

# Display the updated missing values for abbreviation
data_df['Abbreviation'].isna().sum()
```

```python
# Keep rows that only has two missing value at most
data_df = data_df.dropna(thresh=15).reset_index(drop=True)
# Get columns that still contain missing values
null_variables = data_df.isna().sum().reset_index().rename(columns= {0:'total_missing_values'})

# Display null value count
print('Missing value count for each column after dropping values through thresholding')
display(null_variables)
```

```python
# Create a query that locates each columns that has missing values
null_variables = null_variables.loc[null_variables['total_missing_values'] > 0]['index'].tolist()

# Create a temporary copy of the main dataframe to prevent warnings
temp_df = data_df.copy()

# Fill the remaining columns with null values with their median values to make the data less suscpetible to outliers
for variable in null_variables:
  medianValue =temp_df[variable].median()
  temp_df[variable] =temp_df[variable].fillna(round(medianValue,4))

# Display the result
print('\n\nNull value count for each columns after filling missing values with median')
display(temp_df.isna().sum().reset_index().rename(columns= {0:'total_missing_values'}))

# Copy the updated dataframe to the main dataframe
data_df = temp_df.copy()
```

Defining some measures of central tendency and dispersion for aggregate functions:

```python
def mode(x):
    return x.value_counts().idxmax()

def val_range(x):
    return x.max() - x.min()

def interquartile_range(x):
    Q1 = x.quantile(0.25)
    Q3 = x.quantile(0.75)
    return Q3 - Q1

def percentile(n):
    def percentile_(x):
        return x.quantile(n)
    percentile_.__name__ = '{:02.0f}%'.format(n*100)
    return percentile_
```

Import libraries for modelling the data

```python
from sklearn import preprocessing
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
from scipy import stats
import plotly.express as px
```

###Preprocessing

Extract the features to be used from clustering from the main dataframe

```python
observed_columns = ['GDP(in USD)','Co2-Emissions', 'Life expectancy']
observed_df = data_df[observed_columns]
```

Prior to data modelling, the extracted features must be preprocessed in order to improve the accuracy of the model as well as imporve the performance of the model by reducing the computational time. Each numerical features are normalized using L1 Normalization. The reason for using this normalization technique is because it is more robust to outliers as compared to L2 normalization.

```python
scaled_data = preprocessing.normalize(observed_df, norm = 'l1')
observed_df = pd.DataFrame(scaled_data,columns = observed_columns)
```

```python
observed_df.head()
```

In modelling the data, it is important that we choose the optimal number of clusters. The number of clusters is optimally chosen using a method, KElbowVisualizer from the yellowbrick library. This method fits the kMeans model, from sklearn library, a range of K values ranging from 1 to 11. The most optimal number of clusters is then chosen by choosing the point in the visualizer that is annotated with a dashed line.

```python
 # Instantiate the clustering model and visualizer
kmeans = KMeans(random_state=48,n_init='auto')
visualizer = KElbowVisualizer(kmeans, k=(1,11))
visualizer.fit(observed_df.values)
visualizer.show()
```

###Modelling the data

Each datapoint is then clustered with a total of 3 clusters and a max number of 300 iterations to find the optimal clusters of each datapoint.

```python
# Create a KMeans model with 3 number of clusters
kmeans_optimal = KMeans(n_clusters=3, n_init='auto',random_state=48)

# Store the cluster groups in the observed dataframe
observed_df['cluster'] = kmeans_optimal.fit_predict(observed_df)

viz_df = observed_df.copy()
viz_df['cluster'] = viz_df['cluster'].astype(str)
viz_df['GDP(in USD)'] = viz_df['GDP(in USD)']

# Visualize the clusters
colors = ['red', 'green', 'blue']
fig = px.scatter_3d(viz_df, x='Co2-Emissions', y='Life expectancy', z='GDP(in USD)',
              color='cluster',color_discrete_sequence=colors,log_x=True,log_y=True,log_z=True)
fig.show()
```

### **Hypothesis testing:**
**Null Hypothesis:** The mean Minimum wage(in USD) is the same across all clusters.

**Alternative Hypothesis:** At least one cluster has a mean Minimum wage(in USD) that is significantly different from the others.

 **Alpha level to be used will be set at 0.05**

 Testing on Miminum Wage can offer a deeper look into how the clusters intersect environmental and health indicators with one another. Minimum wage serves as an important economic metric that reflect's a nations living standards, and by seeing how this interacts with other variables, it bridges the gap between economic indicators and quality of life.

**CHECKING FOR NORMALITY AND HOMOGENEITY OF VARIANCES**

```python
# Normality check using Shapiro-Wilk test
print('Minimum wage(in USD)')
S_stat, S_p = stats.shapiro(data_df['Minimum wage(in USD)'])
print(f'Shapiro-Wilk test Statistic: {S_stat}, P-Value: {S_p}')
print('\n')

L_stat, L_p = stats.levene(merged_df[merged_df['cluster'] == '0']['Minimum wage(in USD)'],
                       merged_df[merged_df['cluster'] == '1']['Minimum wage(in USD)'],
                       merged_df[merged_df['cluster'] == '2']['Minimum wage(in USD)'])

print(f'Levene’s test Statistic: {L_stat}, P-Value: {L_p}')
```

Based on the Shapiro-Wilk test, we can infer that Minimum Wage is not normally distributed.

Based on the Levene's test, we can also infer that Minimum Wage is not homogeneous across the clusters.

Given the abnormal distribution and non-homogeniety across the clusters, the **Kruskal-Wallis H test** will be used.

```python
stat, p = stats.kruskal(merged_df[merged_df['cluster'] == '0']['Minimum wage(in USD)'],
                       merged_df[merged_df['cluster'] == '1']['Minimum wage(in USD)'],
                       merged_df[merged_df['cluster'] == '2']['Minimum wage(in USD)'])

print(f'Kruskal-Wallis H Test Statistic: {stat}, P-Value: {p}')
```

### **Results:**

The p-value of 0.16974772632912574 is greater than 0.05, meaning there is not enough evidence to reject the null hypothesis at alpha level of 0.05.

This means that there is no statistically significant difference in the median Minimum Wage (in USD) across the three clusters found earlier.
